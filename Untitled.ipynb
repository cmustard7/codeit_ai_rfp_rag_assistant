```python
from src.rag_hf import answer_with_rag, load_store

# (선택) JSON 벡터스토어 미리 로드해서 재사용하면 반복 호출 시 좀 더 빠름
store = load_store()   # Chroma 쓰고 싶으면 생략 가능

question = "최근 1년 이내에 공고된 AI 관련 RFP 중에서, 데이터 구축이 포함된 사업을 간단히 요약해줘."

result = answer_with_rag(
    question,
    store=store,        # JSON vectorstore 사용
    use_chroma=False,   # True로 하면 search_chroma 사용
    top_k=4,
)

print("=== 답변 ===")
print(result["answer"])
print("\n=== 사용된 컨텍스트 메타데이터 ===")
for c in result["contexts"]:
    print(c["metadata"])
```
